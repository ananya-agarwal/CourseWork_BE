```python
#ques 2
plt.plot(sales.month_number,sales.total_profit,'ro--',linewidth=3,markerfacecolor='green',markersize=8)
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.title("Company sales data of last Month")
plt.legend(['Profit data of last year'],loc = "lower right")
plt.show() #iska bina bhi chal jaata ha

#legend me [ ] used varna poora nahi aarha tha
```

```python
#ques 3
plt.plot(sales.month_number,sales.facecream,'bo-')
plt.plot(sales.month_number,sales.facewash,'o-',color='orange')
plt.plot(sales.month_number,sales.toothpaste,'go-')
plt.plot(sales.month_number,sales.bathingsoap,'ro-')
plt.plot(sales.month_number,sales.shampoo,'o-',color = 'purple')
plt.plot(sales.month_number,sales.moisturizer,'o-',color = 'brown')
plt.xlabel("Month number")
plt.ylabel("Sales Unit in Number")
plt.title("Sales data")
plt.legend(['Face Cream Sales Data','Face Wash Sales Data','Tooth Paste Sales Data','Bathing Soap Sales Data','Shampoo Sales Data','Moisturizer Sales Data',])
```

```python
#ques 4
plt.plot(sales.month_number,sales.toothpaste,'o') #by default light blue color
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.legend(['Tooth Paste Sales data'])
plt.title("Tooth Paste Sales data")
plt.grid(linestyle ='--')
```

```python
#ques 5
plt.bar(sales.month_number,sales.facecream,width=0.4,color="blue",label='Face Cream Sales Data')
plt.bar(sales.month_number+0.4,sales.facewash,width=0.4,color="orange",label='Face Wash Sales Data')
plt.xticks(list(range(0,13,1)))
plt.xlabel("Month Number")
plt.ylabel("Sales units in Number")
plt.title("Face wash and Face cream Sales data")
plt.grid(linestyle='--')
plt.legend(loc='upper left')
plt.show()
```

```python
#ques 6

plt.bar(sales.month_number,sales.bathingsoap,width=0.8)
plt.xlabel("Month number")
plt.ylabel("Sales units in Number")
plt.title("bathingsoap sales data")
plt.xticks(list(range(0,13,1)))
plt.grid(linestyle='--')
plt.show()
```

```python
#ques 8
pie_data = [np.sum(sales[col]) for col in sales.columns ]
pie_data
```

```python
pie_data.pop()
pie_data.pop()
pie_data.pop(0)
pie_data #bcs starting ka and last ke 2 columns include nahi krne in the pie chart
```

```python
labels = ['Face Cream','Face Wash','ToothPaste','Bathing Soap','Shampoo','Moisturizer']
plt.pie(pie_data,labels = labels,autopct='%0.1f%%')
#auto  to show % value 
plt.title('Sales Data')
plt.axis('equal') #grapg shift hojaygea(set x and y limits) taaki labels dikhne lage
plt.legend(labels,loc='lower right')
plt.show()
```

```python
#ques 9
plt.subplot(2, 1, 1) #the figure has 2 rows, 1 column and this plot is the first plot.
plt.plot(sales.month_number,sales.bathingsoap,'o-',color='black')
plt.title("Sales data of Bathing Soap")

plt.subplot(2, 1, 2)#the figure has 2 rows, 1 column and this plot is the second plot.
plt.plot(sales.month_number,sales.facewash,'ro-')
plt.title("Sales data of Face wash")

plt.xlabel("Month number")
plt.ylabel("Sales units in number")

plt.show()
```

```python
#ques 10
plt.stackplot(list(range(1,13)),sales.facecream,sales.facewash,sales.toothpaste,sales.bathingsoap,sales.shampoo,sales.moisturizer)
plt.legend(labels,loc='upper left')
plt.xlabel("Month number")
plt.ylabel("Sales Unit in Number")
plt.title("All products sales data using stack plot")
plt.show()
```

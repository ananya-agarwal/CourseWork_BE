```python
data = data.drop(["Model"],axis=1)
print(data)
```

```python
#normalization
for col in data.columns:
    data[col] = data[col]/np.sqrt(np.sum(np.square(data[col])))
```

```python
weight = [1,1,0.5,0.5]
for i,col in enumerate(data.columns):
    data[col]=data[col]*weight[i]
```

```python
data
```

```python
#ideal_best ek array bana say ek row hai jisme optimal(max/min) values padi hai
#(of size 4) jisme corr column ki max value aagyi
#iss row ka har row se euclden dis nikalnge
#Rseq column ki max value aagyi and so on
#sim ideal_worst ek 4 size ka array ban gaya
ideal_best = np.array([max(data.Corr),max(data.Rseq),min(data.RMSE),max(data.Accuracy)])
ideal_worst = np.array([min(data.Corr),min(data.Rseq),max(data.RMSE),min(data.Accuracy)])
```

```python
def Euclidean(row1,row2):
    return np.sqrt(np.sum(np.square(row1-row2)))
```

```python
Splus = np.array([Euclidean(ideal_best,row) for row in data.values ])
Sminus = np.array([Euclidean(ideal_worst,row) for row in data.values ])

data['Splus'] = Splus
data['Sminus'] = Sminus
data
```

```python
PScore = Sminus/(Sminus+Splus)
data['Topsis Score'] = PScore
```

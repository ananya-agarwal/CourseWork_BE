```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
Transposed=X_scaled.T
```

```python
Covariance_matrix=np.cov(Transposed)
```

```python
print(Covariance_matrix)
```

```python
value,vector=np.linalg.eig(Covariance_matrix)
```

```python
print(value)
```

```python
print(vector)
```

```python
percentage_values=[]
for i in range(len(value)):
    percentage_values.append(value[i]/np.sum(value))
print(percentage_values)
```

```python
projected_1=X_scaled.dot(vector.T[0])
projected_2=X_scaled.dot(vector.T[1])
```

```python
res=pd.DataFrame(projected_1,columns=['PC1'])
```

```python
res['PC2']=projected_2
res["Species"]=Y
print(res)
```

```python
from sklearn import datasets
digits=datasets.load_digits()
```

```python
digitdf = pd.DataFrame(data= np.c_[digits['data'], digits['target']], #converting to dataframe
                     columns= digits['feature_names'] + ['target'])
```

```python
digitdf
```

```python
#digiY=digitdf[digitdf.columns[len(digitdf.columns)-1]]
```

```python
digiY=digitdf.iloc[:,-1].values
```

```python
print(digiY)
```

```python
#digiX=digitdf[digitdf.columns[list(range(0,len(digitdf.columns)-1))]]
```

```python
digiX=digitdf.iloc[:,:-1].values
```

```python
print(digiX);
```

```python
digitsX_scaled=scaler.fit_transform(digiX)
```

```python
digitsX_scaled
```

```python
TransposedDigi=digitsX_scaled.T
```

```python
Covariance_matrix_digits=np.cov(TransposedDigi)
```

```python
valuedigi,vectordigi=np.linalg.eig(Covariance_matrix_digits)
```

```python
valuedigi
```

```python
vectordigi
```

```python
percentage_values=[]
for i in range(len(valuedigi)):
    percentage_values.append(valuedigi[i]/np.sum(valuedigi))
print(percentage_values)
```

```python
count=0  #getting required components for accuracy of 95% and above 
sum=0 
for i in percentage_values:
    sum=sum+i
    count=count+1
    if(sum>0.95):
        break
```

```python
count
```

```python
res=pd.DataFrame(digitsX_scaled.dot(vectordigi.T[0]),columns=['PC1']) #making dataframe
sname='PC'
for i in range(42):
    res[sname+str(i+2)]=digitsX_scaled.dot(vectordigi.T[i+1])
```

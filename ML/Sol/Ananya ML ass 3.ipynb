### Binning / discretization: data is first sorted and then the sorted values are distributed into a number of buckets or bins i.e. continous or categorical value ki discrete value banana


### get_dummies converts categorical data into dummy or indicator variables i.e. it performs binning

```python
## birth date (categorical) ko convert in discrete data(age)
## then deleted BirthDate column and added age column
```

```python
df['BirthDate']= pd.to_datetime(df['BirthDate'])

import datetime
CURRENT_TIME = datetime.datetime.now()
def get_age(birth_date,today=CURRENT_TIME):
    y=today-birth_date
    return y.days//365

df['Age']=df['BirthDate'].apply(lambda x: get_age(x))

df.drop(['BirthDate'],axis=1,inplace=True)

df.head()
```

```python
#binning kari ki jaise 5024 places pe ab Partial College nahi..uski jagah pe 1 aayega ab 5024 times
df['Education'].value_counts()
```

```python
df['Education']=df['Education'].map({'Partial High School':1,'High School':2,'Partial College':3,'Bachelors':4,'Graduate Degree':5})
```

```python
print(df['Occupation'].value_counts())
```

```python
df['Occupation']=df['Occupation'].map({'Manual':1,'Skilled Manual':2,'Clerical':3,'Management':4,'Professional':5})
```

```python
df.head()
```

```python
def handle_cardinal_cols(df):
    df['Male']=df['Gender'].map({'M':1,'F':0})
    df.drop(['Gender'],axis=1,inplace=True)
    
    
    df['MaritalStatus']=df['MaritalStatus'].map({'M':1,'S':0})
    
    return df

#1 hi function se 2 works horhe hai..male is mapped to 1 and female to 0
#marital status column me married mapped to 1 and single mapped to 0
```

```python
df=handle_cardinal_cols(df)
df.head()
```

### Binarization (One Hot Encoding): categorical data ko 0/1 mein bana diya

```python
def preprocess_address(df):
    ohe_countries=pd.get_dummies(df['CountryRegionName'],drop_first=True)
    df.drop(columns=['CountryRegionName'],axis=1,inplace=True)
    df=pd.concat([ohe_countries,df],axis=1)
    return df

# CountryRegionName column ki saari rows ko milakar suppose 6 unique values thi say india,pak.... 
# jo jis country ko belong karta tha, vahan 1 value aagayi and rest jagah 0
#..just reduced a column..got 5 new columns now..kyuki CountryRegionName column ko hatadiya and ye 5 columns uski jagah added
```

```python
df=preprocess_address(df)
df.head()
```

### Normalization:  rescales the values into a range of [0,1]
### normalizing the data using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['YearlyIncome','Age']])
df['YearlyIncome_scaled']=scaled[:,0]
df['Age_scaled']=scaled[:,1]
df = df.drop(['YearlyIncome','Age'], axis=1)
df.head()

#scaled list banayi..scaled ka 0 hai YearlyIncome and scaled ka 1 ha Age
#[:0] means scaled ki saari rows and 0 th column YearlyIncome_scaled me aajaaye
```
